### An Exercise on Data Analysis

_*Problem Statement:*_ You are working for an airline company that wants to analyze its passenger data for the past year. The company has provided you with a CSV file containing the monthly passenger data for each airline. Your task is to use NumPy, Pandas, and Matplotlib to analyze the data and answer the following questions:
1) What is the trend of monthly passenger traffic for each airline? Plot the line graph to show the same.

2) What is the overall monthly passenger traffic trend for all airlines combined? Plot the line graph to show the same.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
airlinedata = pd.read_csv('../flights.csv')
```

```python
monthgroupby = airlinedata.groupby('month')['passengers'].sum()
monthgroupby

```

```python
yeargroupby = airlinedata.groupby('year')
yeargroupby.head()
```

```python
generalfigure, axs = plt.subplots(nrows=1, ncols=2, figsize=(12, 8))
monthgroupby.plot(kind='bar', ax=axs[0])
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
axs[1].plot(monthgroupby.index, monthgroupby, marker='o')
axs[1].set_title('Monthly Passengers Over The Years')
axs[1].set_xlabel('Month')
generalfigure.suptitle('Airline Passenger Data Analysis')
plt.show()
```

```python
def factors(x):
    factors = []
    for i in range(1, x+1):
        if x%i == 0:
            factors.append(i)
    return factors

factors(12)

import random
random.randint(2, len(factors(12)))
```

```python
def plots_generator(data):
    numofaxes = len(data)
    rows = random.randint(2, len(factors(numofaxes)))
    fig, axs = plt.subplots(nrows = rows, ncols = int(numofaxes/rows), figsize = (12, 12))
    for ((key, value), ax) in zip(data.items(), axs.flatten()):
        ax.plot((data[key])['month'], (data[key])['month'])


plots_generator(datadict)
```
