```python
class Solution:
    def twoSum(self, nums, target):
        for num in nums:
            for i in range((nums.index(num)) + 1, len(nums)):
                if (num + nums[i]) == target:
                    print(num, "+", nums[i], " = ", num+nums[i])
                    return [nums.index(num), i]
        

obj = Solution()
obj.twoSum([-1,-2,-3,-4,-5], -8)
```
